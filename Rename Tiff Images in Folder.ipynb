```python
import os
import shutil

# Define source and destination folders
source_folder = r"D:\practice\auf1_04242023_msi_chip"  # Change this to your source folder path
destination_folder = r"D:\practice\auf1_04242023_msi_chip_changed"  # Change this to your destination folder path
prefix = "2023_"

# Create destination folder if it does not exist
os.makedirs(destination_folder, exist_ok=True)

# Loop through all files in the source folder
for filename in os.listdir(source_folder):
    if filename.endswith(".tiff") or filename.endswith(".tif"):
        new_filename = prefix + filename
        source_path = os.path.join(source_folder, filename)
        destination_path = os.path.join(destination_folder, new_filename)
        
        # Copy and rename file
        shutil.copy2(source_path, destination_path)
        print(f"Renamed and copied: {filename} -> {new_filename}")

print("All TIFF images have been renamed and saved in the destination folder.")
```
